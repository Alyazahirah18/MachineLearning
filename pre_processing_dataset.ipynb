##Daily Nutrition Dataset

```python
dailynutrition_df = pd.read_csv('AKGbayi.csv')
dailynutrition_df.head()
```

```python
dailynutrition_df.info()
```

```python
dailynutrition_df = dailynutrition_df.astype({'Data.Protein': 'float64', 'Data.Fat.Total Lipid': 'float64', 'Data.Carbohydrate': 'float64', 'Data.Fiber': 'float64', 'Data.Water': 'float64', 'Data.Vitamins.Vitamin E': 'float64', 'Data.Vitamins.Vitamin K': 'float64', 'Data.Niacin': 'float64', 'Data.Choline': 'float64', 'Data.Vitamins.Vitamin C': 'float64', 'Data.Major Minerals.Copper': 'float64'})
```

```python
dailynutrition_df['Data.Major Minerals.Copper'] = dailynutrition_df['Data.Major Minerals.Copper'] / 1000  # Mengubah mcg ke mg
dailynutrition_df['Data.Vitamins.Vitamin E'] = dailynutrition_df['Data.Vitamins.Vitamin E'] / 1000  # Mengubah mcg ke mg
```

```python
dailynutrition_df.info()
```

```python
dailynutrition_df.head()
```

```python
dailynutrition_df.isna().sum()
print("Jumlah duplikasi: ", dailynutrition_df.duplicated().sum())
```
